```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Import excel sheet
xlsx = pd.ExcelFile('StockData.xlsx')
df1 = pd.read_excel(xlsx, 'Sheet2') #2020
df2 = pd.read_excel(xlsx, 'Sheet3') #2019

#Find the volatility of df1
df1["volatility1"] = (df1["Open"] - df1["Open"].mean())**2
sum_column1 = df1.sum(axis=0)
volatilityAvg1 = sum_column1["volatility1"]/253
print("The avg. volatility of 2020 is- ", volatilityAvg1)

# Plot graph
ax2 = df1.plot.line(x = 'Date',
                       y = 'volatility1',
                       c = 'DarkBlue',
                       title='Volatility 2020')
                      



#Find the volatility of df2
df2["volatility2"] = (df2["Open"] - df2["Open"].mean())**2
sum_column2 = df2.sum(axis=0)
volatilityAvg2 = sum_column2["volatility2"]/253
print("The avg. volatility of 2019 is- ", volatilityAvg2)

# Plot graph
ax3 = df2.plot.line(x = 'Date',
                       y = 'volatility2',
                       c = 'DarkBlue',
                       title='Volatility 2019')



# Useful syntax
#print(df1.columns)
#print(df1.info())
#print(df1.shape)
#df1.describe()
```
